## GRÁFICO DE DISPERSIÓN

Muestra la relación entre dos variables continuas.

Se utilizan para mostrar relaciones. 

Para la correlación, los gráficos de dispersión ayudan a mostrar la fuerza de la relación lineal entre dos variables. 

Para la regresión, los gráficos de dispersión suelen incorporar una línea de ajuste. 

En control de calidad, los gráficos de dispersión pueden con frecuencia incluir límites de especificación o líneas de referencia.

```python
import numpy as np
```

```python
x=np.random.rand(100)
y=np.random.rand(100)
```

```python
plt.scatter(x,y, color='orange',alpha=0.8)
plt.title('Gráfico de dispersion', color='indigo', fontsize=20)
```

## DIAGRAMA CAJAS Y BIGOTES
Un diagrama de cajas y bigotes es una manera conveniente de mostrar visualmente grupos de datos numéricos a través de sus cuartiles.

Las líneas que se extienden paralelas a las cajas se conocen como «bigotes», y se usan para indicar variabilidad fuera de los cuartiles superior e inferior. Los valores atípicos se representan a veces como puntos individuales que están en línea con los bigotes. Los diagramas de cajas y bigotes se pueden dibujar vertical u horizontalmente.

```python
fig, axs= plt.subplots(2,2, figsize=(15,10))
#G.LINEAS
axs[0,0].plot(x,y1, label= 'Seno')
axs[0,0].plot(x,y2, label='Coseno')
axs[0,0]. legend(loc= 'upper center')
axs[0,0].set_title('Grafico de líneas', fontsize=15, color='navy')
axs[0,0].grid(color='gold', linestyle='-', linewidth=.5)

#G.BARRAS

axs[0,1].bar(categoria, valores, color='orange', edgecolor='yellow')
axs[0,1].set_title('Grafico de barras', fontsize=15, color='navy')


#G.DISPERSIÓN
axs[1,0].scatter(a,b,color='deeppink', alpha=0.7)
axs[1,0].set_title('Grafico de dispersion',fontsize=15, color='navy')


#HISTOGRAMA
axs[1,1].hist(datos_h,bins=30, color='pink', edgecolor='deeppink') #edgecolor='blue' marca los bordes
axs[1,1].set_title('Histograma',fontsize=15, color='navy')


fig.suptitle('Ejemplos de diferentes tipos de gráficos',fontsize = 20, color='navy')
plt.show()

```
